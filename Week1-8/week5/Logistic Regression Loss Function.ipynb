### Simple plots to understand loss function in Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['legend.fontsize'] = 10
import pandas as pd
```

**Logistic Regression Losses:**

```python
z = np.linspace(-5,5, 200)
```

y=1

```python
cost1 = -1 * np.log(1/(1 + np.exp(-z)))

plt.plot(z, cost1, label='cost1')
plt.xlabel('raw output from hypothesis (theta*X)')
plt.ylabel('cost when y=1')
plt.title("Logistic Regressin Cost when y=1")
plt.legend(["cost y=1"])
plt.show()
```

```python
cost0 = -1 * np.log(1 - (1/(1+np.exp(-z))))

plt.plot(z, cost0, label='cost0')
plt.xlabel('raw output from hypothesis (theta*X)')
plt.ylabel('cost when y=0')
plt.title("Logistic Regressin Cost when y=0")
plt.legend(["cost y=0"])
plt.show()
```

```python
cost0 = -1 * np.log(1 - (1/(1+np.exp(-z))))
cost1 = -1 * np.log(1/(1 + np.exp(-z)))

plt.plot(z, cost0, label='cost, y=0')
plt.plot(z, cost1, label='cost, y=1')
plt.xlabel('raw output from hypothesis (theta*X)')
plt.ylabel('TOTAL COST')
plt.title("Logistic Regressin Cost")
plt.legend()
plt.show()
```
